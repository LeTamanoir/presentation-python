```python
# Chaîne de caractères

c = "abcde"

print(c[0])
print(c[1])
```

```python
# Listes

a = [1, 2.5, 'a', 'b']
print(a[0])

a.append('c')
print(a)

a.insert(0,0)
print(a)

a.pop(4)
print(a)
```

```python
# Opérations

print(1 + 1)
print(5 - 3)
print(3 * 9)
print(10 / 5)
print(32 // 10)
print(32 % 10)
print(2 ** 3)
```

```python
# Structures Itératives

a = 1
b = 1

print(a == b)
print(a != b)

b = 2

print(a > b)
print(a < b)

c1 = True
c2 = False

print(c1 and c2)
print(c1 or c2)


if c1:
  print('hello')
elif c2:
  print('world')
else:
  print('good bye')
```

```python
# Fonctions

def maFonction(x,y):
  return x + y

print(maFonction(1,2))
```

```python
# Boucles

for i in range(4):
  print(i)

print('====')

for i in [1,2,3,4]:
  print(i)

print('====')

i = 0
while i < 10:
  print(i)
  i += 1 # i = i + 1
```
